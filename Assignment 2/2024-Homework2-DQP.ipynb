```python
df["Incident Start Date"] = df["Incident Start Date"].dt.date

df["Incident Start Date"] = pd.to_datetime(df["Incident Start Date"])

df["Incident Start Date"]
```

```python
df.fillna({"Structure Defense Actions Taken" : "Unknown"}, inplace=True)

df["Structure Defense Actions Taken"].describe().T
```

```python
df["AIV missing"] = df["Assessed Improved Value"].isna().astype(int)

# 2. Fill missing with 0
df.fillna({"Assessed Improved Value" : 0}, inplace=True)

# 3. Clamp outliers - values above 90th percentile
percentile_90 = df["Assessed Improved Value"].quantile(0.9)
df["Assessed Improved Value"] = np.minimum(df["Assessed Improved Value"], percentile_90)
```

```python
# 1. Create indicator for missing or invalid (0) years
df["Year Built missing"] = df["Year Built"].isna() | (df["Year Built"] == 0)
df["Year Built missing"] = df["Year Built missing"].astype(int)

# 2. Set 0s to NaN (so they get imputed)
df.loc[df["Year Built"] == 0, "Year Built"] = np.nan

# 3. Impute with median
year_built_median = df["Year Built"].median()
df.fillna({"Year Built": year_built_median}, inplace=True)

# 4. Clamp outliers (e.g., anything before 1800 gets clamped to 1800)
df["Year Built"] = df["Year Built"].clip(lower=1800)
```

```python
# Correcting the column name to handle HTML encoding (&gt;)
df.rename(columns={"Distance - Residence to Utility/Misc Structure &gt; 120 SQFT": "Distance - Residence to Utility/Misc Structure > 120 SQFT"}, inplace=True)

# Remove any single quote from the string and replace '30' with '<30' and '50' with '>50'
df["Distance - Residence to Utility/Misc Structure > 120 SQFT"] = df["Distance - Residence to Utility/Misc Structure > 120 SQFT"].str.replace("'", "")

# Replace '30' with '<30' and '50' with '>50'
df["Distance - Residence to Utility/Misc Structure > 120 SQFT"] = df["Distance - Residence to Utility/Misc Structure > 120 SQFT"].replace({'30': '<30', '50': '>50'})

df.fillna({"Distance - Residence to Utility/Misc Structure > 120 SQFT": "Unknown"}, inplace=True)

df["Distance - Residence to Utility/Misc Structure > 120 SQFT"] = df["Distance - Residence to Utility/Misc Structure > 120 SQFT"].astype("category")

# Check unique values to ensure the changes were applied
df["Distance - Residence to Utility/Misc Structure > 120 SQFT"].describe().T
```

```python
import re

def extract_zip(address):
    if pd.isna(address):
        return None
    match = re.search(r"\b\d{5}\b", str(address))
    return float(match.group(0)) if match else None

df['Zip Code'] = df['Zip Code'].astype('float64')
df.fillna({"Zip Code" : df["Site Address"].apply(extract_zip)}, inplace=True)
df.loc[df["Zip Code"] == 0, "Zip Code"] = df["Site Address"].apply(extract_zip)

zipcode_filter = df[(df["Zip Code"].isnull() != True) & df["Zip Code"] != 0]
df['Zip Code'] = df['Zip Code'].astype('category')

if pd.api.types.is_categorical_dtype(df["Zip Code"]):
    # Add "Unknown" to the existing categories
    df["Zip Code"] = df["Zip Code"].cat.add_categories(["Unknown"])

df.fillna({"Zip Code" : "Unknown"}, inplace=True)

zipcode_filter["Zip Code"] = zipcode_filter["Zip Code"].astype("category")
zipcode_filter["Zip Code"].describe().T
```

```python
# Handle categorical median calculation without integer conversion
def safe_median(series):
    """Calculate median while handling categorical and null values"""
    if pd.api.types.is_numeric_dtype(series):
        return series.median()
    try:
        return pd.to_numeric(series, errors='coerce').median()
    except:
        return np.nan

# Define imputation rules
imputation_rules = {
    'Single Family': 1.0,
    'Mobile Home Single Wide': 1.0,
    'Mobile Home Double Wide': 2.0,
    'Utility': 0.0,
    'Infrastructure': 0.0
}

# Apply direct imputation rules first
for pattern, value in imputation_rules.items():
    mask = (df["Units in Structure"].isna() & 
            df["Structure Type"].str.contains(pattern, case=False, na=False))
    df.loc[mask, "Units in Structure"] = value

# Handle Commercial buildings separately with median
commercial_mask = (df["Units in Structure"].isna() & 
                  df["Structure Type"].str.contains("Commercial", case=False, na=False))

if commercial_mask.any():
    commercial_values = df.loc[df["Structure Type"].str.contains("Commercial", case=False, na=False),
                             "Units in Structure"]
    
    # Calculate median while preserving float dtype
    commercial_median = safe_median(commercial_values)
    
    # Use median if available, otherwise use default 1.0
    fill_value = commercial_median if not np.isnan(commercial_median) else 1.0
    df.loc[commercial_mask, "Units in Structure"] = fill_value

# Fill any remaining missing values with 1.0 (if desired)
df["Units in Structure"] = df["Units in Structure"].fillna(1)
```
