## Feature selection

To analyze the correlation of categorical features, we typically look at how these features relate to one another based on the damage and destruction percentages you have provided. The goal is to identify which features provide distinct and useful information for predicting the outcome.

#### Features Selected (Why We Take Them)
1. CAL FIRE Unit
The percentage of destroyed structures varies greatly between CAL FIRE Units. For example, BTU has a destruction rate of 78%, while BDU is just 3%. This strong variation suggests that the CAL FIRE Unit is highly informative about risk levels, likely due to differences in local fire behavior and suppression capabilities.

2. Structure Type
Structure types show substantial differences in survival â€” mobile homes, especially single and double-wide types, have very high destruction rates (around 70â€“85%), while commercial and multi-story buildings are more resistant. This strong separation makes Structure Type critical.

3. Structure Category
Higher-level grouping into structure categories also provides separation â€” for example, "Single Residence" has about 52% destruction, while "Infrastructure" has only 10%. Including Structure Category adds a useful level of abstraction on top of Structure Type.

4. Roof Construction
Roof type shows noticeable trends: Wood and Asphalt roofs are more likely to be destroyed compared to Concrete or Tile roofs. These differences, aligned with known fire resistance properties, make Roof Construction a valuable feature.

5. Eaves
Structures with enclosed eaves have a much lower destruction rate compared to those with no eaves or unenclosed ones. Since ember intrusion is a major cause of structural loss during wildfires, this architectural detail is important.

6. Vent Screen
Smaller mesh screens (â‰¤ 1/8") show better survival compared to larger screens (>1/8"). Given that vents are a major vulnerability for ember entry, the type of vent screen is a meaningful feature to retain.

7. Exterior Siding
Exterior siding material matters: structures with metal siding perform much better than those with combustible or wood sidings. The variation in destruction percentages across siding types supports its inclusion.

8. Window Pane
Multi-pane windows and "no windows" structures show better survival rates compared to single-pane windows, which are much more likely to fail during a fire. Window construction is a known factor affecting survivability, making this an important predictor.

9. Fence Attached to Structure
Although the separation is not as strong as others, having a combustible fence attached moderately increases destruction likelihood. Given its potential to act as a fire bridge to the structure, this feature adds value.

#### Features Not Selected (Why We Are Not Taking Them)

1. County
CAL Fire unit and County features likely have a geographic or regional correlation. CAL FIRE Units are associated with specific areas, which could overlap with counties. In this case, they might provide redundant information, as both are geographic categorizations. Excluding this.

2. Structure Defense Actions Taken
While certain actions like "Fire Retardant Drop" or "Hand Crew Fuel Break" show very high survival, the "Unknown" category is overwhelmingly large and inconsistent (over 50% destruction). So we exclude it for now.

3. Distance - Propane Tank to Structure
There is some pattern â€” closer propane tanks (0â€“30 feet) have higher destruction rates â€” but again, a huge percentage of unknowns exist (~50%). The predictive gain may not outweigh the cost of handling missing data here.

4. Distance - Residence to Utility/Misc Structure >120 SQFT
Distances less than 30 feet correlate with higher destruction, but again a very large number of "Unknown" entries dilute the strength of this feature. We exclude this feature for now.

5. Deck/Porch (On Grade and Elevated)
Although having a deck does slightly affect destruction probability, the separation is moderate, and unknown values are high. Deck material types also show mixed patterns. Therefore, it is not as impactful as the features we selected.

6. Patio Cover/Carport Attached to Structure
Similarly, presence or absence of a patio cover shows only a mild influence, and the categories have overlapping destruction rates. Its predictive power is relatively weak.

7. Units in Structure
There are some interesting extremes (e.g., 2 units structures have high destruction), but most categories are sparse (very few samples), making the feature noisy and unreliable for general prediction.

#### Some helper functions

1. plot_confusion_matrix(y_true, y_pred, class_names) : Plots a confusion matrix

2. printMetrics(testActualVal, predictions) : Function to get the performance metrics when given the actual values and prediction

3. modelEvaluationMeasures(model, x, y) : Function which takes in a model, the descriptive features and the target features and gets the predictions

```python
def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = metrics.confusion_matrix(y_true, y_pred)
    disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
    
    fig, ax = plt.subplots(figsize=(5, 5))
    disp.plot(cmap='Blues', colorbar=False, ax=ax)
    
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.grid(False)
    plt.show()
```

```python
def printMetrics(testActualVal, predictions):
    # classification evaluation measures
    print('\n==============================================================================')
    print("MAE: ", metrics.mean_absolute_error(testActualVal, predictions))
    print("RMSE: ", metrics.mean_squared_error(testActualVal, predictions)**0.5)
    print("R2: ", metrics.r2_score(testActualVal, predictions))

    print(f"\nAccuracy: {metrics.accuracy_score(testActualVal, predictions):.4f}")
    print(f"\nPrecision: {metrics.precision_score(testActualVal, predictions):.4f}")
    print(f"Recall: {metrics.recall_score(testActualVal, predictions):.4f}")
    print(f"F1 Score: {metrics.f1_score(testActualVal, predictions):.4f}")
    
    print("\nConfusion Matrix:")
    plot_confusion_matrix(testActualVal, predictions, class_names=['Negative', 'Positive'])
```

```python
def modelEvaluationMeasures(model, x, y, linear_regression = False):
    categ_linreg_predictions_train = model.predict(x)

    predicted_classes_train = (categ_linreg_predictions_train >= 0.5).astype(int)

    if linear_regression:
        print("\nPredicted continuous values (first 10 training examples):")
        print(categ_linreg_predictions_train[:10])

        print("\nPredicted classes (after thresholding at 0.5) (first 10 training examples):")
        print(predicted_classes_train[:10])
    else:
        print("\nPredicted classes (first 10 training examples):")
        print(predicted_classes_train[:10])

    printMetrics(y, predicted_classes_train)
```

### **Training the Linear regression model**

As we know, linear regression is mainly used for regression problems, but it can be adapted for classification problems as well.

Let's assume a simple regression problem with one descriptive feature and one target feature. When we plot the points on a graph, the points lie at either y = 0 or y = 1, with varying x-values. The linear regression model tries to minimize the error and draw a line that generally goes diagonally between 0 and 1, with a slope depending on the majority class.

Now, when we give a new input to the model, it outputs a number, which can be anything from -âˆž to +âˆž depending on the x-value. It makes sense to threshold this value at 0.5:

- If the prediction is < 0.5, classify as 0.
- If the prediction is â‰¥ 0.5, classify as 1.

The regression line may underestimate or overestimate, and it may not have high accuracy.

### **Training the Logistic regression model**

Logistic regression is mainly used for classification. It follows the same principle as linear regression, except that the polynomial output is passed through a logistic (sigmoid) function.

Unlike linear regression, the logistic function outputs a number between 0 and 1, representing the probability of class 1 occurring. Considering a simple classification problem with one descriptive feature and one target feature, as x â†’ -âˆž, the probability of class 1 occurring approaches 0, and as x â†’ +âˆž, the probability approaches 1.

Thus, logistic regression naturally fits binary classification tasks better.

### **Training the Random Forest model**

Random Forest is an ensemble learning method that combines multiple decision trees to improve prediction accuracy and reduce overfitting.

Each tree is trained on a random subset of the data (bootstrap sampling) and considers only a random subset of features at each split, introducing diversity.

- For classification, the final prediction is determined by majority voting across all trees.
- For regression, the prediction is determined by averaging the outputs.

This approach enhances robustness against noise and outliers, as individual tree errors tend to cancel each other out. By leveraging randomness in both data and feature selection, Random Forest achieves strong generalization while maintaining interpretability through feature importance scores.

With an increase in tree depth and the number of estimators (i.e., the number of trees in the forest), accuracy tends to increase significantly. However, overfitting can occur â€” the model essentially memorizes the entire dataset with overly specific splitting conditions.
